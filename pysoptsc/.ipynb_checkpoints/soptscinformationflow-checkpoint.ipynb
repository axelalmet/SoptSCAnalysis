## Flows within the same pathway

We will construct contingency tables for/from both samples, to see how we can estimate this.

```python
contingency_tables_samepathway_1 = {}
contingency_proportions_samepathway_1 = {}
cell_clusters = haensel1_so.cell_ids['ClusterName']
cluster_labels = list(cell_clusters.unique())
```

```python
for pathway in flow_paths1:
    aggregate_probability = haensel1_so.aggregated_probabilities[pathway] # Get the aggregated individual cell-cell matrix
    paths = flow_paths1[pathway] # Get all the 3-paths for this pathway
    
    tables = {} # Initialise tables
    proportions = {} # Initialise proportions
    for path in paths:
        table_list = []
        proportion_list = []
        
        # For each triplet (A, B, C), we construct a table the following way:
        #            B -> C    B -/-> C     Total
        # A -> B      K         N - K         N  
        # A -/> B     L         M - L         M
        # where M + N = no. cells of type B
        table = np.zeros((2, 2)) # Initialise table
            
        cluster_A, cluster_B, cluster_C = [cluster_labels[label] for label in path]
        cells_A = np.where(cell_clusters == cluster_A)[0]
        cells_B = np.where(cell_clusters == cluster_B)[0]
        cells_C = np.where(cell_clusters == cluster_C)[0]

        # Calculate the number of cells with an in-edge into B (just > 0 for now.)
        aggregate_probability_AB = aggregate_probability[cells_A, :]
        cells_from_A_to_B = np.intersect1d(aggregate_probability_AB.nonzero()[1], cells_B)
        cells_not_from_A_to_B = np.setdiff1d(cells_B, cells_from_A_to_B)

        # Calculate the number of cells with an out-edge from B(just > 0 for now.)
        aggregate_probability_BC = aggregate_probability[:, cells_C]
        cells_from_B_to_C = np.intersect1d(aggregate_probability_BC.nonzero()[0], cells_B)

        # From this, we can calculate the various intersections of the table
        cells_from_A_to_C = np.intersect1d(cells_from_A_to_B, cells_from_B_to_C) # Cells that go from A to C via B
        cells_not_from_A_to_C = np.intersect1d(cells_not_from_A_to_B, cells_from_B_to_C) # Cells that go from B to C, but not from A

        table[0, 0] = len(cells_from_A_to_C) # These are the number of type B cells with edge from A and edge out to C
        table[0, 1] = len(cells_from_A_to_B) - table[0, 0] # Number of type B cells with edge from A but no edge out to C
        table[1, 0] = len(cells_not_from_A_to_C) # Number of type B cells with no edge from A but edge out to C
        table[1, 1] = len(cells_not_from_A_to_B) - table[1, 0] # Number of type B cells with no edge from A nor no edge to C

        proportion = table.copy()
        proportion = proportion / proportion.sum(axis=1, keepdims=True)
        
        table_list.append(table) # Add the matrix to the list
        proportion_list.append(proportion)
            
        tables[(cluster_A, cluster_B, cluster_C)] = table_list # Set the tables
        proportions[(cluster_A, cluster_B, cluster_C)] = proportion_list # Set the tables
        
    contingency_tables_samepathway_1[pathway] = tables # Define all the contingency tables for the pathways
    contingency_proportions_samepathway_1[pathway] = proportions # Define all the contingency tables for the pathways
```

## Now we need to consider paths between cells across different pathways

```python
pathways = [pathway for pathway in haensel_so.aggregated_cluster_probabilities]
```

```python
pathway_pairs = {(i, j):1 for i, j in itertools.product(pathways, pathways) if i != j }
```

```python
cluster_probability_threshold = 0.05
cell_clusters = list(haensel1_so.cell_ids['ClusterName'].unique())
num_clusters = len(cell_clusters)
node_pairs = {(i, j):1 for i, j in itertools.product(range(num_clusters), range(num_clusters))}

flow_paths_diffpath1 = {}
cluster_probability_nonzero_indices = {}

# We will first store where the (thresholded) cluster probabilities are non-zero, to save us time
for pathway in haensel1_so.aggregated_cluster_probabilities:

    aggregate_cluster_probability = haensel1_so.aggregated_cluster_probabilities[pathway]
    filtered_cluster_probability = aggregate_cluster_probability.copy()
    filtered_cluster_probability[filtered_cluster_probability < cluster_probability_threshold] = 0.0

    cluster_probability_nonzero_indices[pathway] = filtered_cluster_probability.nonzero()


for pathway_pair in pathway_pairs:
    pathway_1, pathway_2 = pathway_pair
    
    paths_between_pair = []
    
    # Get the cell types that exhibit communication activity for each pathway
    sources_1, targets_1 = cluster_probability_nonzero_indices[pathway_1] 
    sources_2, targets_2 = cluster_probability_nonzero_indices[pathway_2]
    
    # We should only consider the pair if there is an intersection between the cell types 
    # in pathway 1 with in-edges and cell types in pathway 2 with out edges
    intermediate_cell_types = np.intersect1d(targets_1, sources_2)
    
    if intermediate_cell_types.any(): # Only consider if there's cell types in common
        
        # For each intermediate cell, we get the start and end poitns
        for cell in intermediate_cell_types:
            
            source_cells = sources_1[np.where(targets_1 == cell)[0]]
            target_cells = targets_2[np.where(sources_2 == cell)[0]]
            
            for source, target in itertools.product(source_cells, target_cells):
                
                paths_between_pair.append([source, cell, target])
    
    if paths_between_pair: # Only add if it's non-empty
        flow_paths_diffpath1[pathway_pair] = paths_between_pair
```

```python
for pathway_pairs in flow_paths_diffpath1:
    pathway_1, pathway_2 = pathway_pairs
    
    # Get the individual cell-cell probabilities
    aggregate_probability1 = haensel2_so.aggregated_probabilities[pathway_1] 
    aggregate_probability2 = haensel2_so.aggregated_probabilities[pathway_2]
    paths = flow_paths_diffpath2[pathway_pairs] # Get all the 3-paths for this pathway
    
    tables = {} # Initialise tables
    table_list = []
    for path in paths:
        
        # For each triplet (A, B, C), we construct a table the following way:
        #            B -> C    B -/-> C     Total
        # A -> B      K         N - K         N  
        # A -/> B     L         M - L         M
        # where M + N = no. cells of type B

        table = np.zeros((2, 2)) # Initialise table

        cluster_A, cluster_B, cluster_C = [cluster_labels[label] for label in path]
        cells_A = np.where(cell_clusters == cluster_A)[0]
        cells_B = np.where(cell_clusters == cluster_B)[0]
        cells_C = np.where(cell_clusters == cluster_C)[0]

        # Calculate the number of cells with an in-edge into B (just > 0 for now.)
        aggregate_probability_AB = aggregate_probability[cells_A, :]
        cells_from_A_to_B = np.intersect1d(aggregate_probability_AB.nonzero()[1], cells_B)
        cells_not_from_A_to_B = np.setdiff1d(cells_B, cells_from_A_to_B)

        # Calculate the number of cells with an out-edge from B(just > 0 for now.)
        aggregate_probability_BC = aggregate_probability[:, cells_C]
        cells_from_B_to_C = np.intersect1d(aggregate_probability_BC.nonzero()[0], cells_B)

        # From this, we can calculate the various intersections of the table
        cells_from_A_to_C = np.intersect1d(cells_from_A_to_B, cells_from_B_to_C) # Cells that go from A to C via B
        cells_not_from_A_to_C = np.intersect1d(cells_not_from_A_to_B, cells_from_B_to_C) # Cells that go from B to C, but not from A

        table[0, 0] = len(cells_from_A_to_C) # These are the number of type B cells with edge from A and edge out to C
        table[0, 1] = len(cells_from_A_to_B) - table[0, 0] # Number of type B cells with edge from A but no edge out to C
        table[1, 0] = len(cells_not_from_A_to_C) # Number of type B cells with no edge from A but edge out to C
        table[1, 1] = len(cells_not_from_A_to_B) - table[1, 0] # Number of type B cells with no edge from A nor no edge to C
            
        tables[(cluster_A, cluster_B, cluster_C)] = table # Set the tables
    contingency_tables_diffpathway_2[pathway_pairs] = tables # Define all the contingency tables for the pathways
```

### Calculate the contingency tables for the combined samples

```python
cluster_probability_threshold = 0.05
cell_clusters = list(haensel_so.cell_ids['ClusterName'].unique())
num_clusters = len(cell_clusters)
node_pairs = {(i, j):1 for i, j in itertools.product(range(num_clusters), range(num_clusters))}

flow_paths_samepathway = {}
for pathway in haensel_so.aggregated_cluster_probabilities:

    aggregate_cluster_probability = haensel_so.aggregated_cluster_probabilities[pathway]
    filtered_cluster_probability = aggregate_cluster_probability.copy()
    filtered_cluster_probability[filtered_cluster_probability < cluster_probability_threshold] = 0.0

    # Should re-normalise the probabilities
    # We now divide each non-zero row by the sums
    nonzero_rows = filtered_cluster_probability.nonzero()[0]
    filtered_cluster_probability[nonzero_rows, :] /= filtered_cluster_probability[nonzero_rows, :].sum(axis=1, keepdims=True) 

    cluster_network = nx.from_numpy_array(filtered_cluster_probability, create_using = nx.DiGraph())

    three_paths = []
    for pair in node_pairs:
        i, j = pair
        paths = [list(path) for path in nx.all_simple_paths(cluster_network, source = i, target = j, cutoff = 2) if len(path) > 2]
        if paths:
            for path in nx.all_simple_paths(cluster_network, source = i, target = j, cutoff = 2):
                if len(path) > 2: # Should be a three-path
                    three_paths.append(path)
                    
    flow_paths_samepathway[pathway] = three_paths
```

```python
cluster_probability_threshold = 0.05
cell_clusters = list(haensel_so.cell_ids['ClusterName'].unique())
num_clusters = len(cell_clusters)
node_pairs = {(i, j):1 for i, j in itertools.product(range(num_clusters), range(num_clusters))}

flow_paths_diffpath = {}
cluster_probability_nonzero_indices = {}

# We will first store where the (thresholded) cluster probabilities are non-zero, to save us time
for pathway in haensel_so.aggregated_cluster_probabilities:

    aggregate_cluster_probability = haensel_so.aggregated_cluster_probabilities[pathway]
    filtered_cluster_probability = aggregate_cluster_probability.copy()
    filtered_cluster_probability[filtered_cluster_probability < cluster_probability_threshold] = 0.0

    cluster_probability_nonzero_indices[pathway] = filtered_cluster_probability.nonzero()


for pathway_pair in pathway_pairs:
    pathway_1, pathway_2 = pathway_pair
    
    paths_between_pair = []
    
    # Get the cell types that exhibit communication activity for each pathway
    sources_1, targets_1 = cluster_probability_nonzero_indices[pathway_1] 
    sources_2, targets_2 = cluster_probability_nonzero_indices[pathway_2]
    
    # We should only consider the pair if there is an intersection between the cell types 
    # in pathway 1 with in-edges and cell types in pathway 2 with out edges
    intermediate_cell_types = np.intersect1d(targets_1, sources_2)
    
    if intermediate_cell_types.any(): # Only consider if there's cell types in common
        
        # For each intermediate cell, we get the start and end poitns
        for cell in intermediate_cell_types:
            
            source_cells = sources_1[np.where(targets_1 == cell)[0]]
            target_cells = targets_2[np.where(sources_2 == cell)[0]]
            
            for source, target in itertools.product(source_cells, target_cells):
                
                paths_between_pair.append([source, cell, target])
    
    if paths_between_pair: # Only add if it's non-empty
        flow_paths_diffpath[pathway_pair] = paths_between_pair
```
